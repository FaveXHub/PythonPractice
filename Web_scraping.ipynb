```python
import requests
page=requests.get(url)
```

```python
BeautifulSoup(page.text,'html')
```

```python
#assign html
soup=BeautifulSoup(page.text,'html')
print(soup)
```

```python
soup.find_all('table')
```

```python
table=soup.find_all('table')
```

```python
soup.find('table',last_='https://companiesmarketcap.com/')
```

```python
print(table)
```

```python
top_companies=soup.find('table',class_='marketcap-table')
```

```python
top_companies=soup.find_all('th')
print(top_companies)
```

```python
#use list comprehension and clean
top_largest_companies=[title.text.strip() for title in top_companies]
print(top_largest_companies)
```

```python
import pandas as pd
```

```python
df=pd.DataFrame(top_largest_companies)
```

```python
df=pd.DataFrame(columns=top_largest_companies)
df
```

```python
column_data=table[0].find_all('tr')
```

```python
# Loop through each row in the table body (skipping the header row)
for row in column_data[1:]:
    # Find all table data cells (td) in the current row
    row_data = row.find_all('td')
    # Extract the text from each data cell, strip whitespace, and store in a list
    individual_row_data = [data.text.strip() for data in row_data]
    # Print the extracted data for the current row
    print(individual_row_data)
```

```python
# Loop through each row in the table body (skipping the header row)
for row in column_data[1:]:
    # Find all table data cells (td) in the current row
    row_data = row.find_all('td')
    # Extract the text from each data cell, strip whitespace, and store in a list
    individual_row_data = [data.text.strip() for data in row_data]
    # Check if the number of columns matches the DataFrame before appending
    if len(individual_row_data) == len(df.columns):
        # Append the extracted data to the DataFrame
        df.loc[len(df)] = individual_row_data

# Display the populated DataFrame
display(df)
```

```python
df.to_csv(r'/content/largest_companies.csv', index=False)
```
